```python
orders=df.groupby(['order_id'])['material'].apply(list)
```

Составим словарь пар, для быстрого доступа к количеству совместных встреч (в одном заказе) для данной пары по паре материалов

Также составим таблицу, в которой ключи это id материалов, а значения списки, в которых:
1) список уникальных пар в которых встречался данный материал
2) число, обозначающее в скольких заказах присутствовал данный товар (материал)

```python
cooc_dict = {}
materials = {}
for order in tqdm.tqdm_notebook(orders.values):
    order = np.unique(order)
    for i in range(0,len(order)):
        id = order[i]
        if id not in materials.keys():
            materials[id] = [[],1]
        elif id in materials.keys():
            materials[id][1]+=1                     
        for j in range(i,len(order)):
            id2 = order[j]
            if id2 != id:
                key = str(id2) + '_' + str(id)
                if key in cooc_dict.keys():
                    cooc_dict[key] += 1
                elif (str(id) + '_' + str(id2)) in cooc_dict.keys():
                    key = str(id) + '_' + str(id2)
                    cooc_dict[key] += 1
                else:
                    cooc_dict[key] = 1
                    materials[id][0].append(key)
            
```

```python
answ2 = pd.DataFrame()
for org in np.unique(df['org']):
    df2 = df[df['org']==org]
    orders=df2.groupby(['order_id'])['material'].apply(list)

    cooc_dict = {}
    materials = {}
    for order in tqdm.tqdm_notebook(orders.values):
        order = np.unique(order)
        for i in range(0,len(order)):
            id = order[i]
            if id not in materials.keys():
                materials[id] = [[],1]
            elif id in materials.keys():
                materials[id][1]+=1                     
            for j in range(i,len(order)):
                id2 = order[j]
                if id2 != id:
                    key = str(id2) + '_' + str(id)
                    if key in cooc_dict.keys():
                        cooc_dict[key] += 1
                    elif (str(id) + '_' + str(id2)) in cooc_dict.keys():
                        key = str(id) + '_' + str(id2)
                        cooc_dict[key] += 1
                    else:
                        cooc_dict[key] = 1
                        materials[id][0].append(key)
    unique_materials = np.unique(df2['material'])
            
    count = 0
    basic = []
    related = []
    for id in unique_materials:
        try:
          max_cooc = cooc_dict[materials[id][0][np.argmax([cooc_dict[i] for i in materials[id][0]])]]
        except:
          max_cooc = 0
        if max_cooc/materials[id][1]>0.6 and max_cooc>10:
          count+=1
          related.append(str(id))
        else:
          basic.append(str(id))

    bas = pd.Series(basic, name = 'Осн товары_' + str(org))
    rel = pd.Series(related, name = 'Сопутствующие_' + str(org))
    answ2 =pd.concat([answ2, bas, rel], axis = 1)
```
